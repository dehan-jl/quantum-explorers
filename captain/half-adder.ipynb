```python
from qiskit import QuantumCircuit

qcs = {}
for i in range(4):
    qc = QuantumCircuit(4, 2)
    q0 = i % 2
    q1 = (i // 2) % 2
    if q0 == 1:
        qc.x(0)
    if q1 == 1:
        qc.x(1)
    qc.cx(0, 2)
    qc.cx(1, 2)
    qc.ccx(0, 1, 3)
    qc.measure([2, 3], [0, 1])
    qc.draw()
    qcs[f'{q1}{q0}'] = qc

for id, qc in qcs.items():
    print(f'{id}')
    display(qc.draw(output='mpl'))
```

```python
from qiskit import Aer
sim = Aer.get_backend('aer_simulator')

for id, qc in qcs.items():
    print(f'{id} -> ', end='')
    print(list(sim.run(qc).result().get_counts().keys())[0])
```
